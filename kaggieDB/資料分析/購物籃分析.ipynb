### 移除購物籃分析不需要的欄位

```python
df1 = df[["BASKET_ID","PRODUCT_ID","QUANTITY"]]
```

```python
df1
```

### 排除QUANTITY為0的資料

```python
df2 = df1.query('QUANTITY > 0')
```

```python
df2
```

```python
product_df.head()
```

```python
df3 = pd.merge(df2,product_df,on='PRODUCT_ID')
```

```python
df3.head()
```

```python
df3 = df3.drop('PRODUCT_ID',axis=1)
```

```python
df3
```

```python
df4 = df3.groupby(["BASKET_ID","SUB_COMMODITY_DESC"])["QUANTITY"].sum()
```

```python
df4
```

```python
df4 = df3.groupby(["BASKET_ID","SUB_COMMODITY_DESC"])["QUANTITY"].sum().unstack()

```

```python
df4.head()
```

```python
df4 = df3.groupby(["BASKET_ID","SUB_COMMODITY_DESC"])["QUANTITY"].sum().unstack().reset_index().fillna(0).set_index('BASKET_ID')
```

```python
df4.head()
```

```python
def encode_units(x):
    if x <= 0:
        return 0
    if x >= 1:
        return 1
```

```python
df5 = df4.applymap(encode_units)
```

```python
df6 = df5[(df5 > 0).sum(axis=1)>=2]
```

```python
df6.head()
```

```python
from mlxtend.frequent_patterns import apriori
```

```python
frequent_itemsets_plus = apriori(df6, min_support=0.03,use_colnames=True).sort_values('support',ascending=False).reset_index(drop=True)
```

```python
frequent_itemsets_plus.head()
```

```python
from mlxtend.frequent_patterns import association_rules
```

```python
association_rules(frequent_itemsets_plus,metric='lift',min_threshold=1).sort_values('lift',ascending=False).reset_index(drop=True)
```
